```python
import eazyml
import pandas as pd
```

```python
#Authentication
auth_token = eazyml.ez_auth('audreygallier@outlook.com', 'hih1hIh!')["token"]
```

```python
options = {
    
    "id" : "job_id",
    "impute" : "no",
    "outcome" : "fraudulent",
    "discard" : ["department", "salary_range", "benefits"]

}

load_response = eazyml.ez_load(auth_token, 'fake_job_postings.xlsx', options)
print(load_response)
```

```python
df_id = load_response['dataset_id']
model = eazyml.ez_init_model(auth_token, df_id)
```

```python
model_id = model['model_id']
new_model = eazyml.ez_build_models(auth_token, model_id)
```

```python
prediction_df = eazyml.ez_predict(auth_token, 12465, 'fake_job_postings.xlsx')
print(prediction_df)
```
